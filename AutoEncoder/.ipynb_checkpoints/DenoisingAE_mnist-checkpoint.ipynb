```python
import numpy as np
import tensorflow as tf
import tensorflow.keras as K
import matplotlib.pyplot as plt
```

```python
SEED = 1926
np.random.seed(SEED)
tf.random.set_seed(SEED)
```

```python
#hyperparametes
BATCH_SIZE = 256
MAX_EPOCHS = 50
LEARNING_RATE = 1e-3
MOMENTUM = 8e-1
HIDDEN_DIM = 1000  #enerally bigger than input size for DAE
ORIGINAL_DIM = 784
OPTIMIZER = tf.keras.optimizers.Adam(learning_rate = LEARNING_RATE)
VERBOSE = 1
```

```python
#load data from MNIST dataset and preprocess
(x_train, _), (x_test, _) = K.datasets.mnist.load_data()
x_train = x_train / 255.
x_test = x_test / 255.
x_train = x_train.astype(np.float32)
x_test = x_test.astype(np.float32)
x_train = np.reshape(x_train, (x_train.shape[0], ORIGINAL_DIM))
x_test = np.reshape(x_test, (x_test.shape[0], ORIGINAL_DIM))
```

```python
#create noise to corrupt inputs
noise_train = np.random.normal(loc=0.5, scale=0.5, size=x_train.shape)
x_train_noisy = x_train + noise_train
noise_test = np.random.normal(loc=0.5, scale=0.5, size=x_test.shape)
x_test_noisy = x_test + noise_test
#saturate x between 0 and 1
x_train_noisy = np.clip(x_train_noisy, 0., 1.)
x_test_noisy = np.clip(x_test_noisy, 0., 1.)
```

```python
#istantiate vanilla autoencode
from Autoencoders import AutoEncoder
autoencoder = AutoEncoder(hidden_dim = HIDDEN_DIM, original_dim = ORIGINAL_DIM)
```

```python
#let's train the denoising autoencoder using Keras inbuilt functions
autoencoder.compile(loss = 'mse', optimizer = OPTIMIZER)
history = autoencoder.fit(x_train_noisy, x_train, validation_data = (x_test_noisy, x_test), epochs = MAX_EPOCHS, batch_size = BATCH_SIZE, verbose = VERBOSE)
```

```python
autoencoder.loss = history.history['loss']
```

```python
#plot loss vs epochs
autoencoder.plot_loss()
```

```python
#### PLOT REAL VS RECONSTRUCTED IMAGES ###
autoencoder.plot_real_vs_reconstructed(data = x_test_noisy)
```
